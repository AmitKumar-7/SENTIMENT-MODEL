# 🎯 Multimodal Meme Sentiment Analysis - VS Code Testing

This notebook allows you to test your multimodal sentiment analysis system interactively in VS Code.

## 🚀 Features to Test:
- **Text Analysis** - Sentiment from text content
- **Image Analysis** - OCR + Visual sentiment + Image captioning
- **Auto-Detection** - Automatic input type detection
- **Performance Testing** - Speed and accuracy measurements


```python
# 🔧 Initialize the System
import os
import json
import time
from datetime import datetime
from advanced_multimodal_analyzer import AdvancedMultimodalAnalyzer

print("🔄 Initializing Multimodal Sentiment Analyzer...")
analyzer = AdvancedMultimodalAnalyzer()
print("✅ System Ready!")
```

## 📝 Text Analysis Testing
Test how the system analyzes text-based memes and sentiment.

```python
# 📝 Test Text Analysis
def test_text(text):
    """Test text sentiment analysis"""
    print(f"📝 Testing: \"{text}\"")
    print("-" * 50)
    
    result = analyzer.analyze_meme(text, "text")
    
    sentiment = result.get("sentiment", "Unknown")
    confidence = result.get("confidence", 0)
    emoji = {"Positive": "😊", "Negative": "😞", "Neutral": "😐"}.get(sentiment, "❓")
    
    print(f"🎯 Result: {emoji} {sentiment} ({confidence:.1%} confidence)")
    
    if "all_scores" in result:
        print("\n📊 Detailed Scores:")
        for label, score in result["all_scores"].items():
            bar = "█" * int(score * 20)
            print(f"   {label:8}: {score:.1%} {bar}")
    
    return result

# Test various text samples
test_texts = [
    "This meme is absolutely hilarious! 😂🤣",
    "I hate this stupid meme",
    "This is just a regular meme about cats",
    "LMAO this made my day!",
    "This meme is cringe"
]

for text in test_texts:
    test_text(text)
    print()
```

## 📸 Image Analysis Testing
Test the multimodal image analysis capabilities.

```python
# 📸 Test Image Analysis
def test_image(image_path):
    """Test image sentiment analysis"""
    print(f"🖼️  Testing: {image_path}")
    print("-" * 50)
    
    if not os.path.exists(image_path):
        print(f"❌ Image not found: {image_path}")
        return None
    
    result = analyzer.analyze_meme(image_path, "image")
    
    if "error" in result:
        print(f"❌ Error: {result['error']}")
        return result
    
    sentiment = result["final_sentiment"]
    confidence = result.get("final_confidence", 0)
    emoji = {"Positive": "😊", "Negative": "😞", "Neutral": "😐"}.get(sentiment, "❓")
    
    print(f"🎯 Final Result: {emoji} {sentiment} ({confidence:.1%} confidence)")
    
    breakdown = result.get("analysis_breakdown", {})
    
    if breakdown.get("extracted_text"):
        text = breakdown["extracted_text"][:100].replace('\n', ' ')
        print(f"📝 Extracted Text: \"{text}...\"")
    
    if breakdown.get("image_caption"):
        print(f"🖼️  Image Caption: \"{breakdown['image_caption']}\"")
    
    print("\n🔍 Analysis Breakdown:")
    if breakdown.get("text_sentiment") and "sentiment" in breakdown["text_sentiment"]:
        ts = breakdown["text_sentiment"]
        print(f"   📝 Text: {ts['sentiment']} ({ts['confidence']:.1%})")
    
    if breakdown.get("visual_sentiment") and "sentiment" in breakdown["visual_sentiment"]:
        vs = breakdown["visual_sentiment"]
        print(f"   👁️  Visual: {vs['sentiment']} ({vs['confidence']:.1%})")
    
    if "meme_categories" in result:
        print(f"🏷️  Categories: {', '.join(result['meme_categories'])}")
    
    return result

# Test available images
if os.path.exists("meme_dataset"):
    image_files = [f for f in os.listdir("meme_dataset") if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    print(f"Found {len(image_files)} images in meme_dataset")
    
    # Test first 3 images
    for image_file in image_files[:3]:
        image_path = os.path.join("meme_dataset", image_file)
        test_image(image_path)
        print()
else:
    print("❌ No meme_dataset folder found")
```

## 🔄 Auto-Detection Testing
Test the system's ability to automatically detect input types.

```python
# 🔄 Test Auto-Detection
def test_auto_detection(input_data):
    """Test auto-detection of input type"""
    print(f"🔄 Auto-Detection Test: {input_data}")
    print("-" * 50)
    
    result = analyzer.analyze_meme(input_data)  # No type specified
    
    detected_type = result.get("input_type", "unknown")
    print(f"✅ Detected Type: {detected_type}")
    
    if "final_sentiment" in result:
        sentiment = result["final_sentiment"]
        confidence = result.get("final_confidence", 0)
    else:
        sentiment = result.get("sentiment", "Unknown")
        confidence = result.get("confidence", 0)
    
    emoji = {"Positive": "😊", "Negative": "😞", "Neutral": "😐"}.get(sentiment, "❓")
    print(f"🎯 Result: {emoji} {sentiment} ({confidence:.1%} confidence)")
    
    return result

# Test auto-detection
test_inputs = [
    "This is a funny text meme!",
    "meme_dataset/positive_1.jpg" if os.path.exists("meme_dataset/positive_1.jpg") else "No image available"
]

for input_data in test_inputs:
    if "No image available" not in input_data:
        test_auto_detection(input_data)
        print()
```

## ⚡ Performance Testing
Measure the speed and efficiency of the analysis.

```python
# ⚡ Performance Testing
def performance_test(input_data, input_type):
    """Test performance with timing"""
    print(f"⚡ Performance Test: {input_type.title()}")
    print("-" * 30)
    
    start_time = time.time()
    result = analyzer.analyze_meme(input_data, input_type)
    end_time = time.time()
    
    processing_time = end_time - start_time
    print(f"✅ Processing Time: {processing_time:.2f} seconds")
    
    if "final_sentiment" in result:
        sentiment = result["final_sentiment"]
        confidence = result.get("final_confidence", 0)
    else:
        sentiment = result.get("sentiment", "Unknown")
        confidence = result.get("confidence", 0)
    
    print(f"✅ Result: {sentiment} ({confidence:.1%})")
    
    return processing_time, result

# Run performance tests
performance_results = []

# Text performance
text_time, text_result = performance_test("This meme is hilarious!", "text")
performance_results.append(("Text", text_time))
print()

# Image performance (if available)
if os.path.exists("meme_dataset/positive_1.jpg"):
    image_time, image_result = performance_test("meme_dataset/positive_1.jpg", "image")
    performance_results.append(("Image", image_time))

# Performance summary
print("\n📊 Performance Summary:")
print("-" * 30)
for test_type, time_taken in performance_results:
    print(f"{test_type:8}: {time_taken:.2f}s")
```
